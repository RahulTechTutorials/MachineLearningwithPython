# Technique 1 - Check the percentage of missing values -- 
Perc missing > 95% - Drop the column
              95 < % Missing < 50 - No imputation required.
              % Missing < 50 - impute

```python
desc = df.describe().T # transpose the describe dataframe and check for count column
```

```python
desc
```

```python
##generate another column which will check the columns having missing %
desc['%Missing'] = 1 - ( desc['count'])/len(df)
```

```python
desc
```

```python
drop_column_with_high_missing_val = []
for index,missing in desc['%Missing'].items():
    if missing > .95:
        drop_column_with_high_missing_val.append(index)
        
```

```python
drop_column_with_high_missing_val
```

```python
df.drop(drop_column_with_high_missing_val,axis=1,inplace=True)
```
