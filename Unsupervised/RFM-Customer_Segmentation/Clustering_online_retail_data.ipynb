```python
import numpy as np
```

```python
def score(x,series, recency=False):
    if not(recency):
        if x <= series.quantile(0.2):
            return 1
        elif x <= series.quantile(0.4):
            return 2
        elif x <= series.quantile(0.6):
            return 3
        elif x <= series.quantile(0.8):
            return 4
        else:
            return 5
    else :
        if x <= series.quantile(0.2):
            return 5
        elif x <= series.quantile(0.4):
            return 4
        elif x <= series.quantile(0.6):
            return 3
        elif x <= series.quantile(0.8):
            return 2
        else:
            return 1
        
```

```python
df7 = df6.copy(deep=True)
df7["F_score"] = df7.Frequency.map(lambda x : score(x,df7.Frequency))
df7["M_score"] = df7.Monetary.map(lambda x : score(x,df7.Monetary))
df7["R_score"] = df7.Recency.map(lambda x : score(x,df7.Recency,True))
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters= 3, random_state=22, max_iter= 1000,verbose=0)
cluster.fit(df7[["F_score","M_score","R_score"]])
```

```python
labels = cluster.predict(df7[["F_score","M_score","R_score"]])
df7['labels'] = labels
df7.labels.value_counts()
```

```python
from sklearn.metrics import silhouette_score
```

```python
s_score = silhouette_score(df7[["F_score","M_score","R_score"]],labels)
print('score',s_score)
```

```python
for i in range(2,30):
    cluster = KMeans(n_clusters= i, random_state=22, max_iter= 1000,verbose=0)
    cluster.fit(df7[["F_score","M_score","R_score"]])
    labels = cluster.predict(df7[["F_score","M_score","R_score"]])
    s_score = silhouette_score(df7[["F_score","M_score","R_score"]],labels)
    print('K: ',i,'| score: ',s_score)
    
```
