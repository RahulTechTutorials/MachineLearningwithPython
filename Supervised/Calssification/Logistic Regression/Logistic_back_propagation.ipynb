# Logistic Regression - Back Propagation

```python
import numpy as np
class Logistic(object):
    def __init__(self, no_of_cols = 1, no_of_rows = 100, learning_rate = 0.01, epochs = 100):
        self.no_of_cols = no_of_cols
        self.no_of_rows = no_of_rows
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_cols)
        self.epochs = epochs
        
    def sigmoid(self, x,deriv=False):
        '''Will give 1/(1+exp(-x)) and derivation'''
        if deriv==True:
            result = x * (1- x)
        else:
            result = 1/(1+np.exp(-x))
        return result
    def equation(self,x):
        '''To calculate mx+c'''
        return np.dot(x,self.weights)
    def predict(self,x):
        '''will give probability'''
        return self.sigmoid(self.equation(x))

    def fit(self,training_inputs, labels):
        for i in range(self.epochs):
            print("Epoch: ", i+1)
            print("weights: ",self.weights)
            for inputs , label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                gradient = self.sigmoid(prediction,True)
                delta = (label-prediction) * gradient
                self.weights += self.learning_rate * delta * inputs       
```
